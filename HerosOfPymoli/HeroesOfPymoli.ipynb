```python
#Import Modules
import pandas as pd

#Load CSV file
file = "purchase_data.csv"

#Read CSV File and store into Pandas data frame
pymoli_df = pd.read_csv(file)
```

```python
#Player Count
#Find number of unique screen names
total_players = pymoli_df["SN"].nunique()

#display total number of players in dataframe
total_players_df = pd.DataFrame({"Total Players" :[total_players]})

total_players_df
```

```python
#Purchasing analysis (Total): finding values
total_uitems = pymoli_df["Item ID"].nunique()
avg_price = pymoli_df['Price'].mean()
num_purchase = pymoli_df["Purchase ID"].count()
total_rev = pymoli_df["Price"].sum()
```

```python
#Purchasing Analysis (Total): Creating dictionary 
purchase_dict = {
                    "Number of Unique Items" : [(total_uitems)], "Average Price" : [(avg_price)],
                    "Number of Purchases" : [(num_purchase)], "Total Revenue" : [(total_rev)]}

#Create Dataframe with formatting
purchase_df = pd.DataFrame(purchase_dict)

purchase_df["Average Price"]= purchase_df["Average Price"].map("${:.2f}".format)
purchase_df["Total Revenue"] = purchase_df["Total Revenue"].map("${:,.2f}".format)

purchase_df
```

```python
#Gender Demographics
#Grouping by Gender
gender_group_df = pymoli_df.groupby(["Gender"])

#Creating dataframe
gender_df = pd.DataFrame()

gender_df['Total Count'] = gender_group_df['SN'].nunique()
gender_df['Percentage of Players'] = gender_df['Total Count']/total_players *100

#format column 
gender_df["Percentage of Players"]= gender_df["Percentage of Players"].map("{:.2f}%".format)

#Sort high to low
gen_order = gender_df.sort_values( 'Total Count', ascending = False)

gen_order





```

```python
#Purchasing Analysis (Gender)
#Still using groupby gender from above line

#Creating dataframe
gen_pur_df = pd.DataFrame()

gen_pur_df["Purchase Count"] = gender_group_df["Purchase ID"].count()
gen_pur_df["Average Purchase Price"] = gender_group_df["Price"].mean()
gen_pur_df["Total Purchase Price"] = gender_group_df["Price"].sum()
gen_pur_df["Avg Total Purchase per Person"] = gender_group_df["Price"].sum() / gender_df['Total Count']

#Format columns
gen_pur_df["Average Purchase Price"]= gen_pur_df["Average Purchase Price"].map("${:.2f}".format)
gen_pur_df["Total Purchase Price"]= gen_pur_df["Total Purchase Price"].map("${:,.2f}".format)
gen_pur_df["Avg Total Purchase per Person"]= gen_pur_df["Avg Total Purchase per Person"].map("${:.2f}".format)

gen_pur_df
```

```python
#Age Demographics
#Binning

# Create bins 
bins = [0,10,15,20,25,30,35,40,150]

# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]
```

```python
#Age Demographics
#Add column with bins to dataframe

pymoli_df ['Age Range'] = pd.cut(pymoli_df['Age'], bins, labels = group_labels, include_lowest = True, right = False)

#Group by newly created bins of Age Range
age_group_df = pymoli_df.groupby(["Age Range"])

#Create dataframe
age_df = pd.DataFrame()
age_df['Total Count'] = age_group_df['SN'].nunique()
age_df['Percentage of Players'] = age_df['Total Count']/total_players *100

#format column
age_df["Percentage of Players"]= age_df["Percentage of Players"].map("{:.2f}%".format)

age_df
```

```python
#Purchasing Analysis (Age)
#Still using groupby Age Range from above line

#Create Dataframe
age_pur_df = pd.DataFrame()

age_pur_df["Purchase Count"] = age_group_df["Purchase ID"].count()
age_pur_df["Average Purchase Price"] = age_group_df["Price"].mean()
age_pur_df["Total Purchase Price"] = age_group_df["Price"].sum()
age_pur_df["Avg Total Purchase per Person"] = age_group_df["Price"].sum() / age_df['Total Count']

#Format columns
age_pur_df["Average Purchase Price"]= age_pur_df["Average Purchase Price"].map("${:.2f}".format)
age_pur_df["Total Purchase Price"]= age_pur_df["Total Purchase Price"].map("${:,.2f}".format)
age_pur_df["Avg Total Purchase per Person"]= age_pur_df["Avg Total Purchase per Person"].map("${:.2f}".format)

age_pur_df
```

```python
#Top Spenders

#Group by Screen Name
sn_group_df = pymoli_df.groupby(["SN"])

#Create dataframe
spend_df = pd.DataFrame()

spend_df["Purchase Count"] = sn_group_df["SN"].count()
spend_df["Average Purchase Price"] = sn_group_df["Price"].mean()
spend_df["Total Purchase Value"] = sn_group_df["Price"].sum()

#Find 5 top spenders
high_spend_df = spend_df.nlargest(5,"Total Purchase Value")


#Format columns
high_spend_df["Average Purchase Price"]= high_spend_df["Average Purchase Price"].map("${:.2f}".format)
high_spend_df["Total Purchase Value"]= high_spend_df["Total Purchase Value"].map("${:,.2f}".format)


high_spend_df
```

```python
#Most Popular Items
#Group by Item ID and Item Name
item_group_df = pymoli_df.groupby(["Item ID","Item Name"])

#Create dataframe
item_df = pd.DataFrame()

item_df["Purchase Count"] = item_group_df["SN"].count()
item_df["Item Price"] = item_group_df["Price"].sum() / item_df['Purchase Count']
item_df["Total Purchase Value"] = item_group_df["Price"].sum()

#Find the 5 most popular items
high_item_df = item_df.nlargest(5,"Purchase Count", keep = "last")


#Format columns
high_item_df["Item Price"]= high_item_df["Item Price"].map("${:.2f}".format)
high_item_df["Total Purchase Value"]= high_item_df["Total Purchase Value"].map("${:,.2f}".format)


high_item_df
```

```python
#Most Profitable Items
#Still using Group by Item ID and Item Name from above line

#Create dataframe
item_df = pd.DataFrame()

item_df["Purchase Count"] = item_group_df["SN"].count()
item_df["Item Price"] = item_group_df["Price"].sum() / item_df['Purchase Count']
item_df["Total Purchase Value"] = item_group_df["Price"].sum()

#Find the 5 most profitable items
high_item_df = item_df.nlargest(5,"Total Purchase Value", keep = "last")


#Format columns
high_item_df["Item Price"]= high_item_df["Item Price"].map("${:.2f}".format)
high_item_df["Total Purchase Value"]= high_item_df["Total Purchase Value"].map("${:,.2f}".format)


high_item_df
```
